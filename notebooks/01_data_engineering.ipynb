```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import sweetviz as sv
import os

sns.set_theme(style="whitegrid")

```

```python
HOUSING_PATH = "/home/fidisroxy/development/mlops/house-pred-mlops/data/raw/"
def load_housing_data(housing_path=HOUSING_PATH):
    csv_path = os.path.join(housing_path, "house_prices.csv")
    return pd.read_csv(csv_path)
```

```python
housing = load_housing_data()
```

```python
housing.shape
```

```python
housing.info()
```

```python
unwanted_string_labels= ['Super Area']


for label in unwanted_string_labels:
    
    housing[label] = pd.to_numeric(housing[label])
```

```python
int_labels_strored_as_string = ['Bathroom', 'Balcony']

for label in int_labels_strored_as_string:
    housing[label] = housing[label].str.replace(r"\D+","", regex=True)
    housing[label] = pd.to_numeric(housing[label])
```
